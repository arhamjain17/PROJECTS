```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("zomato.csv")
df1.head()
```

```python
df2=pd.read_csv("Copy of Country-Code.csv")
df2.head()
```

# 1)	Merge the country sheet and Zomato file to get the country code


```python
df=pd.merge(df1,df2,how="inner",on="Country Code")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

# 2) Get the Top rated restaurants in each city in India(On the basis of rating and votes).

### By Rating 

```python
a=df[df["Country"]=="India"]
b=a["City"].value_counts()
c=a.sort_values(by="Aggregate rating",ascending=False)
for i in b.index:
    x=a[a["City"]==i]
    print(i)
    print(x[["Restaurant Name","Aggregate rating"]][x["Aggregate rating"]==x["Aggregate rating"].max()])
    print()
```

### By Votes 

```python
a=df[df["Country"]=="India"]
b=a["City"].value_counts()
c=a.sort_values(by="Votes",ascending=False)
for i in b.index:
    x=a[a["City"]==i]
    print(i)
    print(x[["Restaurant Name","Votes"]][x["Votes"]==x["Votes"].max()])
    print()
```

# 3)	Finding out relationship between the rating and votes.

```python
a=df.groupby("Aggregate rating",as_index=False)["Votes"].sum()
plt.figure(figsize=(30,20))
ax=sns.barplot(data=a,x="Aggregate rating",y="Votes")

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

```python
a=df.groupby("Aggregate rating",as_index=False)["Votes"].mean()
plt.figure(figsize=(30,20))
ax=sns.barplot(data=a,x="Aggregate rating",y="Votes")

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

```python
plt.figure(figsize=(30,20))
sns.lineplot(data=df,y="Votes",x="Aggregate rating")
plt.show()
```

```python
plt.figure(figsize=(30,20))
sns.scatterplot(data=df,y="Votes",x="Aggregate rating")
plt.show()
```

10)	Top Percentage cover of restaurants in a city.(Draw pie graph for this).

```python
x=df.groupby(["Country","City"],as_index=False).size().sort_values(by="size",ascending=False)
x.head(10)
```

```python
z=x["Country"].value_counts()
for i in z.index:
    top=x[x["Country"]==i][["City","size"]].head(5)
    plt.title(i)
    plt.pie(top["size"],labels=top["City"],autopct='%1.1f%%', startangle=140)
    plt.show()
```
